```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pathlib
import numpy as np
```

```python
file_path = pathlib.Path(r'C:\Users\juann\Notebook\Formula 1\retail_sales_data.csv')
```

```python
df = pd.read_csv(file_path)
```

```python
df.columns
```

```python
df.info()
```

# Sales amount insights

# Stacked bar displaying sales amount per month, through 2022 and 2023

```python
sales_pivot_table = pd.pivot_table(df,values='Sales Amount',
                                   index=['Year', 'Month'], aggfunc="sum",
                                   sort=False).unstack(level=0)
# Display the updated DataFrame with flattened column names
sales_pivot_table.columns = ['_'.join([str(i) for i in col if i]) for col in sales_pivot_table.columns]

sales_pivot_table.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.title('Stacked Bar Plot of Sales Amount for 2022 and 2023')
plt.ylabel('Sales Amount')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.tight_layout()
plt.show()
plt.close()
```

# Barplot comparing sales amount per month, through 2022 and 2023

```python
sales_pivot_table.plot(kind='bar', figsize=(10, 6))
plt.title('Sales Amount for 2022 and 2023')
plt.ylabel('Sales Amount')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```

# Two bar plots illustrating the Sales Amount for the years 2022 and 2023.

```python
sales_amount = sales_pivot_table.reset_index()


# Creating the subplots
fig, ax = plt.subplots(2, 1, figsize=(10, 8))

# Plotting for 2022
ax[0].bar(sales_amount['Month'],
          sales_amount['Sales Amount_2022'],
          color='skyblue')
ax[0].set_title('Sales Amount in 2022')
ax[0].set_ylabel('Sales Amount')

# Plotting for 2023
ax[1].bar(sales_amount['Month'],
          sales_amount['Sales Amount_2023'],
          color='lightgreen')
ax[1].set_title('Sales Amount in 2023')
ax[1].set_ylabel('Sales Amount')

# Adjust layout
plt.tight_layout()
plt.show()
```

# Sales Growth YoY

```python
# Pivot data to have years as columns with months as the index and total sales as values
pivot_sales = df.pivot_table(values='Sales Amount', index='Month', columns='Year', aggfunc='sum') # pass an index to mantain order

# Ensure the months are sorted correctly for comparison
months_order = ["January", "February", "March", "April", "May", "June", 
                "July", "August", "September", "October", "November", "December"]
# reidex
pivot_sales = pivot_sales.reindex(months_order)

# Formula: (2023 sales - 2022 sales) / 2022 sales * 100
pivot_sales['Growth Rate (%)'] = ((pivot_sales[2023] - pivot_sales[2022]) / pivot_sales[2022]) * 100

# Visualization for Monthly Growth Comparison
plt.figure(figsize=(10, 6))
pivot_sales['Growth Rate (%)'].plot(kind='bar', color='skyblue')
plt.title('Monthly Growth Rate from 2022 to 2023')
plt.ylabel('Growth Rate (%)')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Display the calculated growth rates
print("Monthly Growth Rate from 2022 to 2023:")
print(pivot_sales['Growth Rate (%)'])
```

```python
# Splitt the data on Year
sales_2022 = df[df['Year'] == 2022]
sales_2023 = df[df['Year'] == 2023]

# Sum sales per month given year
sales_2022_sum = sales_2022.groupby(['Month'], sort=False).agg({'Sales Amount':'sum'})
sales_2023_sum = sales_2023.groupby(['Month'], sort=False).agg({'Sales Amount':'sum'})

# # Ensure the months are sorted correctly for comparison
# months_order = ["January", "February", "March", "April", "May", "June", 
#                 "July", "August", "September", "October", "November", "December"]
# sales_2022_sum = sales_2022_sum.reindex(months_order)
# sales_2023_sum = sales_2023_sum.reindex(months_order)
```

# Cummulative Sales Comparison

```python
sales_2022_cumsum = sales_2022_sum.cumsum()
sales_2023_cumsum = sales_2023_sum.cumsum()

# Visualization for Cumulative Sales Analysis
plt.figure(figsize=(10, 6))
plt.plot(sales_2022_cumsum, marker='o', label='2022', color='blue')
plt.plot(sales_2023_cumsum, marker='o', label='2023', color='green')
plt.title('Cumulative Sales Comparison: 2022 vs 2023')
plt.ylabel('Cumulative Sales Amount')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

# Sales Difference per Month

```python
# Sales Difference per Month
sales_difference = sales_2022_sum - sales_2023_sum

# Visualization for Sales Difference per Month
plt.figure(figsize=(10, 6))
sales_difference.plot(kind='bar', color='orange')
plt.title('Sales Difference per Month (2023 - 2022)')
plt.ylabel('Sales Difference')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.close()
```

```python
# Average Monthly Sales
average_monthly_sales_2022 = sales_2022_sum.mean()
average_monthly_sales_2023 = sales_2023_sum.mean()


# print("\n4. Average Monthly Sales:")
print("Average Monthly Sales:")
print(f"2022: {average_monthly_sales_2022}")
print(f"2023: {average_monthly_sales_2023}")
```

# Customer Rating Insights

```python
# category with the most customer rating
rating = df.groupby(['Year','Product Category']).agg({'Customer Rating': 'mean'}).reset_index()

# Creating the subplots
fig, ax = plt.subplots(2, 1, figsize=(10, 8))

# Plotting for 2022
ax[0].bar(rating[rating['Year']==2022]['Product Category'],
          rating[rating['Year']==2022]['Customer Rating'],
          color='skyblue')
ax[0].set_title('Sales Amount in 2022')
ax[0].set_ylabel('Sales Amount')

# Plotting for 2023
ax[1].bar(rating[rating['Year']==2023]['Product Category'],
          rating[rating['Year']==2023]['Customer Rating'],
          color='lightgreen')
ax[1].set_title('Sales Amount in 2023')
ax[1].set_ylabel('Sales Amount')

# Adjust layout
plt.tight_layout()
plt.show()
```
